```python
import pandas as pd
```

```python
data={'Age': [36,42,23,52,43,44,66,35,52,35,24,18,45],'Experience':[10,12,4,4,21,14,3,14,13,5,3,3,9],
      'Rank':[9,4,6,4,8,5,7,9,7,9,5,7,9],'Nationality':['UK','USA','N','USA','USA','UK','N','UK','N','N','USA','UK','UK'],
      'GO':['NO','NO','NO','NO','YES','NO','YES','YES','YES','YES','NO','YES','YES']}
df=pd.DataFrame(data)
df
```

```python
import pandas as pd
from sklearn import tree
import pydotplus
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
import pydotplus
from pydotplus import graphviz

%matplotlib inline
```

```python
d={'UK':0,'USA':1,'N':2}
df['Nationality']=df['Nationality'].map(d)
```

```python
d1={'YES':1,'NO':0}
df['GO']=df['GO'].map(d1)
```

```python
features=['Age','Experience','Rank','Nationality']
```

```python
x=df[features]
y=df['GO']
```

# APPLYING DECISION TREE

```python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
```

```python
dtree=dtree.fit(x,y)
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(60,30))
plot_tree(dtree, filled=True, feature_names=features)
```

```python
pred=dtree.predict(x)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y,pred))
```

```python
print(classification_report(y,pred))
```
