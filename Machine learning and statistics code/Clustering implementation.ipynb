```python
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
Data={"a":[25,34,22,27,33,33,31,22,35,34,67,54,57,43,50,57,59,52,65,47,49,48,35,33,44,45,38,43,51,46],
      'b':[79,51,53,78,59,74,73,57,69,75,51,32,40,47,53,36,35,58,59,50,25,20,14,12,20,5,29,27,8,7]}
```

```python
df=DataFrame(Data,columns=['a','b'])
```

```python
kmeans=KMeans(n_clusters=3)
```

```python
kmeans.fit(df)
```

```python
centroids=kmeans.cluster_centers_
```

```python
y_kmeans=kmeans.predict(df)
```

```python
plt.scatter(Data['a'],Data['b'],c=y_kmeans)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],c='red',marker='x')
```

# Example 2..Elbow point

```python
from  sklearn.datasets import make_blobs
```

#make_blobs is used to genertae sample points around c centers (randomly choose)

```python
X,y=make_blobs(n_samples=100,centers=2,cluster_std=1,n_features=2)
```

```python
plt.scatter(X[:,0],X[:,1],s=30,color='b')
plt.xlabel("X")
plt.ylabel("y")
plt.show()
plt.clf() #clear the figure
cost=[]
for i in range(1,11):
    KM=KMeans(n_clusters=i,max_iter=500)
    KM.fit(X)
#calculates squared error for the clusterd points
    cost.append(KM.inertia_)
plt.plot(range(1,11),cost,color='g',linewidth='3')
plt.xlabel("Value of K")
plt.ylabel("Squred Error(cost)")
plt.show()
```
